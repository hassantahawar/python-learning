```python
df = pd.DataFrame({
    "English": [67, 87, 90],
    "Maths": [87, 57, 70],
    "Computer": [77, 71, 40],
}, index=["Hassan", "Musa", "Esha"])
```

```python
df
```

```python
def add_grace_marks(x):
    if x < 60:
        return x + 5
    else:
        return x
```

```python
df.Maths = df.Maths.apply(add_grace_marks)
```

```python
df["Stats"] = df.apply(lambda row: 90 if row["Maths"] > 85 else 78, axis=1)
```

```python
df
```

```python
df.loc["sum", :] = [1,1,1,1]
```

```python
df = df.drop("total", axis=0)
```

```python
df.loc["total", :] = [df["English"].sum(), df["Maths"].sum(), df["Maths"].sum(), df["Stats"].sum()]
```

```python
df["Job"] = ["Programmer", "Programmer", "Designer"]
```

```python
df
```

```python
df.groupby("Job").agg({
    "Maths": ["mean", "min", "max", "count"]
})
```
